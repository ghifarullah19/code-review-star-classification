# **Stars Classification**

This is a dataset consisting of several features of stars.

Some of them are:

- Absolute Temperature (in K)
- Relative Luminosity (L/Lo)
- Relative Radius (R/Ro)
- Absolute Magnitude (Mv)
- Star Color (white,Red,Blue,Yellow,yellow-orange etc)
- Spectral Class (O,B,A,F,G,K,,M)
- Star Type **(Red Dwarf, Brown Dwarf, White Dwarf, Main Sequence , SuperGiants, HyperGiants)**
- Lo = 3.828 x 10^26 Watts (Avg Luminosity of Sun)
- Ro = 6.9551 x 10^8 m (Avg Radius of Sun)

Program ini bertujuan untuk mengklasifikasikan Spektral Kelas Bintang K dengan 6 kelas lainnya.

```python
# Code berikut untuk mengimport library yang dibutuhkan, yaitu pandas dan numpy
import pandas as pd
import numpy as np
```

```python
# Code berikut untuk membaca dataset yang akan digunakan, yaitu dataset bintang yang diambil dari link berikut
star = pd.read_csv('https://github.com/YBIFoundation/Dataset/raw/main/Stars.csv')
```

```python
# Code berikut untuk menampilkan 5 data teratas dari dataset
star.head()
```

```python
# Code berikut untuk menampilkan informasi dari dataset
star.info()
```

```python
# Code berikut untuk menampilkan semua kolom yang ada pada dataset
star.columns
```

```python
# Code berikut untuk menampilkan deskripsi dari dataset, yaitu berupa statistik deskriptif dari kolom numerik dataset
star.describe()
```

```python
# Code berikut untuk menghitung jumlah data pada kolom 'Spectral Class'
star[['Spectral Class']].value_counts()
```

Spektral Kelas dipetakkan dengan angka 1 jika kelasnya adalah A, B, F, O, K, G dan 0 jika kelasnya adalah M

```python
# Code berikut untuk mengubah kolom 'Spectral Class' menjadi kode numerik
star.replace({'Spectral Class':{'M':0, 'A':1, 'B':1, 'F':1, 'O':1, 'K':1, 'G':1 }}, inplace=True)
```

```python
# Code berikut untuk menghitung jumlah data pada kolom 'Star type'
star[['Star type']].value_counts()
```

Star color dipetakkan dengan angka 0 jika warnanya adalah Red, 1 jika warnanya adalah Yellow, 2 jika warnanya adalah White, dan 3 jika warnanya adalah Blue.

```python
# Code berikut untuk mengubah kolom 'Star color' menjadi kode numerik
star.replace({'Star color':{ 'Red':0, 'Yellow':1, 'White':2, 'White ': 2, 'Blue ':3, 'Blue':3 }}, inplace=True)
```

```python
# Code berikut untuk menghitung jumlah data pada kolom 'Star color'
star[['Star color']].value_counts()
```

Variabel dependen atau yang akan diuji adalah kolom Spectral Class sedangkan variabel independen atau yang mempengaruhi variabel dependen adalah kolom Temperature, Luminosity, Radius, dan Absoute magnitude.

```python
# Code berikut memilih kolom yang akan digunakan sebagai variabel dependen (y) dan independen (X)
y = star['Spectral Class']
X = star[['Temperature (K)', 'Luminosity (L/Lo)', 'Radius (R/Ro)',
       'Absolute magnitude (Mv)']]
```

Data yang dilatih dan diuji dibagi menjadi 80:20 dengan random_statenya adalah 200.

```python
# Code berikut untuk membagi dataset menjadi data training dan data testing
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size = 0.8, random_state = 200)
```

```python
# Code berikut untuk membuat model machine learning
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
```

```python
# Code berikut untuk melatih model machine learning
model.fit(X_train,y_train)
```

```python
# Code berikut untuk membuat prediksi menggunakan model machine learning
y_pred = model.predict(X_test)
```

```python
# Code berikut untuk menampilkan hasil prediksi
y_pred
```

Hasil prediksi model menujukkan bahwa model memprediksi semua data sebagai 1 (kelas A, B, F, O, K, G) karena jumlah data kelas 1 lebih banyak dari kelas 0 (M).

```python
# import function
from sklearn.metrics import confusion_matrix, classification_report
```

```python
# Code berikut untuk menampilkan confusion matrix
confusion_matrix(y_test,y_pred)
```

Dari confusion matrix, kita bisa melihat bahwa model kita memiliki akurasi yang cukup baik dengan nilai 14 dan 4 untuk True Positive dan True Negative serta 1 dan 29 untuk False Positive dan False Negative

```python
# Code berikut untuk menampilkan classification report
print(classification_report(y_test,y_pred))
```

Hasil dari classification report menunjukkan bahwa model memiliki akurasi sebesar 0.90 dan memiliki f1-score sebesar 0.85 dan 0.92 untuk kelas 0 dan 1.
